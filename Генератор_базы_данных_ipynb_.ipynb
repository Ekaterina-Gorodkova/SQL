```python
import numpy as np
import faker
import datetime
```

# clients

```python
names = ['Jacob', 'Emily', 'Michael', 'Olivia', 'William', 'Sophia', 'James', 'Ava', 'Benjamin', 'Isabella', 'Elijah', 'Mia', 'Daniel', 'Charlotte', 'Alexander', 'Amelia', 'Matthew', 'Harper', 'Jackson', 'Evelyn']
```

```python
CLIENTS_NUMBER = 30
```

```python
fake = faker.Faker()
start_date = datetime.date(year=1950, month=1, day=1)
end_date = datetime.date(year=2005, month=1, day=1)

for client_id in range(CLIENTS_NUMBER):
    date = fake.date_between(start_date=start_date, end_date=end_date)
    print(f"({client_id}, '{np.random.choice(names)}', '{np.random.choice(names)}', '{date.strftime('%Y-%m-%d')}'),")
```

# orders

```python
fake = faker.Faker()
start_date = datetime.date(year=2023, month=4, day=1)
end_date = datetime.date(year=2023, month=6, day=30)

for order_id in range(100):
    client_id = np.random.randint(CLIENTS_NUMBER + 1)
    items_amount_in_cheque = np.random.randint(1, 30)
    for _ in range(items_amount_in_cheque):
        amount = np.random.randint(1, 100)
        item_id = np.random.randint(150)
        date = fake.date_between(start_date=start_date, end_date=end_date)
        print(f"({client_id}, {order_id}, {item_id}, {amount}, '{date.strftime('%Y-%m-%d')}'),")
```
